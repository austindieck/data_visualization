```python
from bs4 import BeautifulSoup
import requests
import csv
import urllib.request
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

url = 'https://www2.census.gov/programs-surveys/popest/datasets/2010-2020/state/totals/nst-est2020-alldata.csv'

df = pd.read_csv(url, sep=',')

index0 = df.columns.get_loc('NAME')
index1 = df.columns.get_loc('POPESTIMATE2010')
index2 = df.columns.get_loc('POPESTIMATE2020')

df1 = df.loc[:, 'POPESTIMATE2010':'POPESTIMATE2020']
df1.insert(0, 'Years', df['NAME'])
df1.set_index('Years', inplace = True)
df2 = df1.T
print(df2.head())

plt.figure(figsize=(20, 20))
sns.set_palette('Blues')
sns.set_style('darkgrid')
sns.lineplot(data=df2)
sns.set_context('poster')
plt.xticks(rotation=90)
plt.legend(loc='best', fontsize=10)
plt.show()


```
